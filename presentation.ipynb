```python
from network3 import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

dataset = Dataset("mnist")

epoch = 9

k1 = np.load("weights/mnist/k1_" + str(epoch) + "_mnist.npy")
k2 = np.load("weights/mnist/k2_" + str(epoch) + "_mnist.npy")
f1 = np.load("weights/mnist/f1_" + str(epoch) + "_mnist.npy")
f2 = np.load("weights/mnist/f2_" + str(epoch) + "_mnist.npy")

network = ConvNetwork("mnist",k1,k2,f1,f2)

shap_map = network.shap_pixel(dataset.X_train[0],dataset.y_train[0],dataset,20)

im = plt.imshow(shap_map.reshape((21,21)),cmap="coolwarm",vmin=np.min(shap_map),vmax=np.max(shap_map))
```

```python
network = ConvNetwork("mnist",k1,k2,f1,f2)
shap_map_superpixel = network.shap_superpixel(dataset.X_train[0],dataset.y_train[0],dataset,20)
```

```python
fig = plt.figure(frameon=False)
plt.imshow(shap_map_superpixel.reshape((21,21)),cmap="coolwarm")
plt.imshow(dataset.X_train[0].reshape((21,21)),alpha=0.2,cmap="Greys")
```

```python
cat_dataset = Dataset("catsdogs")
catk1 = np.load("weights/catsdogs/k1" + str(epoch) + ".npy")
catk2 = np.load("weights/catsdogs/k2" + str(epoch) + ".npy")
catf1 = np.load("weights/catsdogs/f1" + str(epoch) + ".npy")
catf2 = np.load("weights/catsdogs/f2" + str(epoch) + ".npy")

network = ConvNetwork("catsdogs",catk1,catk2,catf1,catf2)
shap_map_superpixel_cats = network.shap_superpixel(cat_dataset.X_train[0],cat_dataset.y_train[0],cat_dataset,2)
```

```python
fig = plt.figure(frameon=False)
plt.imshow(shap_map_superpixel_cats.reshape((100,100)),cmap="coolwarm",vmin=np.min(shap_map_superpixel_cats),vmax=np.max(shap_map_superpixel_cats),alpha=0.5)
plt.imshow(cat_dataset.X_train[0].reshape((100,100,3)),alpha=0.5)
```

```python
shap_map_clusters_cats = network.shap_cluster(cat_dataset.X_train[0],cat_dataset.y_train[0],cat_dataset,2,2)
```

```python
figure, ax = plt.subplots(1,2,figsize=(10,10))

ax[0].imshow(shap_map_clusters_cats.reshape((100,100)),cmap="coolwarm")
ax[1].imshow(cat_dataset.X_train[0].reshape((100,100,3)))
```
