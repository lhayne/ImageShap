```python
cat_dataset = Dataset("catsdogs")
k1 = np.load("weights/catsdogs/k1" + str(24) + ".npy")
k2 = np.load("weights/catsdogs/k2" + str(24) + ".npy")
f1 = np.load("weights/catsdogs/f1" + str(24) + ".npy")
f2 = np.load("weights/catsdogs/f2" + str(24) + ".npy")

network = ConvNetwork("catsdogs",k1,k2,f1,f2)
```

```python
shap_map_superpixel_cats = network.shap_superpixel(cat_dataset.X_train[0],cat_dataset.y_train[0],cat_dataset,2)
```
