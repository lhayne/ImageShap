```python
from network3 import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

dataset = Dataset("mnist")

epoch = 9

k1 = np.load("weights/mnist/k1_" + str(epoch) + "_mnist.npy")
k2 = np.load("weights/mnist/k2_" + str(epoch) + "_mnist.npy")
f1 = np.load("weights/mnist/f1_" + str(epoch) + "_mnist.npy")
f2 = np.load("weights/mnist/f2_" + str(epoch) + "_mnist.npy")

network = ConvNetwork("mnist",k1,k2,f1,f2)

shap_map = network.shap_pixel(dataset.X_train[3],dataset.y_train[3],dataset,20)

im = plt.imshow(shap_map.reshape((21,21)),cmap="coolwarm",vmin=np.min(shap_map),vmax=np.max(shap_map))
```
